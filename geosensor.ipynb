```python
#removing outliers
q1=df['mean'].quantile(0.25)
q3=df['mean'].quantile(0.75)
iqr=q3-q1
lower=q1-1.5*iqr
upper=q3-1.5*iqr
data=df[(df['mean']>=lower) &(df['mean']<=upper)]
print(data)
```
