## importing MinMaxScalar library for fitting numerical columns

```python
scaler=MinMaxScaler()
df[numerical_cols]=scaler.fit_transform(df[numerical_cols])
df
```

## finding outliers

```python

plt.figure(figsize=(9,9))
sns.boxplot(data=df)
plt.title("outliers in features")
plt.show()
```

## Removing outliers from Outlier column(Volume)

```python

Q1 = df['Volume'].quantile(0.35)
Q3 = df['Volume'].quantile(0.45)
IQR = Q3 - Q1


lower_bound = Q1 - 1.35 * IQR
upper_bound = Q3 + 1.45 * IQR
filtered_data = df[(df['Volume'] >= lower_bound) & (df['Volume'] <= upper_bound)]

plt.subplot(1, 2, 1)
sns.boxplot(x=df['Volume'], color='yellow')
plt.title('Before Removing Outliers')

plt.subplot(1, 2, 2)
sns.boxplot(x=filtered_data['Volume'], color='red')
plt.title('After Removing Outliers')
plt.tight_layout()
plt.show()
```
