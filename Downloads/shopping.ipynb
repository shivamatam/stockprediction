# importing libraries 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
shopping=pd.read_csv('shopping_trends.csv')
shopping.head()
```

```python
shopping.isnull()
```

```python
shopping.describe
```

# visualizing purchaseditem using Histplot

```python
plt.figure(figsize=(20,5))
sns.histplot(shopping['Item Purchased'])
plt.show()
```

```python
size_counts = shopping['Size'].value_counts(normalize=True) * 100
size_percentages = size_counts.reset_index()
size_percentages.columns = ['Size', 'Item Purchased']
```

```python
#for bivariate data points we use the scatterplot or kde
plt.figure(figsize=(10,5))
#plt.plot(shopping['Size'],shopping['Item Purchased'], linestyle='-')
sns.barplot(data=size_percentages, x='Size', y='Item Purchased')
#sns.lineplot(data=shopping,x='Size',y='Item Purchased',color='red')
plt.title('purchased items  Size')
plt.xlabel('Size')
plt.ylabel('Item Purchased')
plt.show()
```

```python
plt.figure(figsize=(20,5))
sns.lineplot(data=shopping,x='Item Purchased',y='Purchase Amount (USD)',color='red')
sns.set_theme(style="darkgrid")
plt.title('purchased items')
plt.xlabel('Item Purchased')
plt.ylabel('Purchase Amount (USD)')
plt.show()
```

```python
sns.displot(shopping, x="Purchase Amount (USD)",y="Item Purchased",col="Gender",color='orange')
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.relplot(
    shopping,
    x="Purchase Amount (USD)", y="Item Purchased", hue="Location", col="Gender")
plt.show()
```
